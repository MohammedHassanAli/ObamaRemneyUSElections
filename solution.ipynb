```python
avg = pd.DataFrame(poll_df.mean())
avg.drop('Number of Observations',inplace=True)
```

```python
avg
```

```python
std = pd.DataFrame(poll_df.std())
std.drop('Number of Observations' , inplace= True)
```

```python
std
```

```python
 avg.plot(yerr=std,kind='bar',legend=False)
```

```python
poll_avg = pd.concat([avg,std],axis=1)
```

```python
poll_avg.head()
```

```python
poll_avg.columns =  {'Average' , 'STD'}
```

```python
poll_df.plot(x='End Date' , y=['Obama','Romney','Undecided'],linestyle='',marker='o')
plt.xticks(rotation='vertical')
```

```python
poll_df['Differance'] = (poll_df['Obama'] - poll_df['Romney']) / 100
```

```python
poll_df.head()
```

```python
poll_df_grouped = poll_df.groupby('Start Date',as_index = False).mean()
```

```python
poll_df_grouped
```

```python
poll_df_grouped.plot('Start Date' , 'Differance' , figsize = (15 , 4) ,marker = 'o' , linestyle='-')
```

```python
row_in = 0
xlimit = []

for date in poll_df_grouped['Start Date']:
    if date[0:7] == '2012-10':
        xlimit.append(row_in)
    row_in += 1 


print(min(xlimit))
print(max(xlimit))
```
